### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

# Data for the bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']  # Custom colors for each bar

# Create the bar plot
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color=colors, alpha=0.8)

# Customizations
plt.title("Bar Plot of Categories vs. Values")  # Title
plt.xlabel("Categories")  # X-axis label
plt.ylabel("Values")  # Y-axis label
plt.grid(axis='y', linestyle="--", alpha=0.6)  # Add horizontal grid lines

# Show the plot
plt.show()
```
