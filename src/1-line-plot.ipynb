### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)  # 100 points for smooth curve
y = x**2  # Square of x

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="y = x²", color="blue", linewidth=2)

# Customizations
plt.title("Line Plot of y = x²")  # Title
plt.xlabel("X-axis (0 to 10)")  # X-axis label
plt.ylabel("Y-axis (Square of X)")  # Y-axis label
plt.grid(True, linestyle="--", alpha=0.7)  # Add grid with dashed lines
plt.legend()  # Show legend

# Show the plot
plt.show()

```
