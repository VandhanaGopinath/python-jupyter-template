### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the histogram
data = np.random.randn(500)  # 500 random numbers from a normal distribution

# Create the histogram
plt.figure(figsize=(7, 5))
plt.hist(data, bins=20, color="teal", edgecolor="black", alpha=0.7)  # Custom color & edges

# Customizations
plt.title("Customized Histogram")  # Title
plt.xlabel("Value Range")  # X-axis label
plt.ylabel("Frequency")  # Y-axis label
plt.grid(axis="y", linestyle="--", alpha=0.6)  # Add horizontal grid lines

# Show the plot
plt.show()
```
