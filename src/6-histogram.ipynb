### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt
import numpy as np


data = np.random.randn(500) 


if len(data) == 0:
    print("Error: Histogram data is empty!")
elif not isinstance(data, (list, np.ndarray)):
    print("Error: Data must be a list or numpy array!")
else:
    
    plt.figure(figsize=(7, 5))
    plt.hist(data, bins=20, color="teal", edgecolor="black", alpha=0.7)  

    
    plt.title("Customized Histogram")  
    plt.xlabel("Value Range") 
    plt.ylabel("Frequency")  
    plt.grid(axis="y", linestyle="--", alpha=0.6)  

    
    plt.show()

```
