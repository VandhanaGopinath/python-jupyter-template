### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt


labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]


if not labels or not sizes:
    print("Error: Data is missing!")
elif len(labels) != len(sizes):
    print("Error: Labels and sizes must have the same length!")
elif any(s < 0 for s in sizes):
    print("Error: Slice sizes cannot be negative!")
else:
    
    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, shadow=True, explode=[0.1, 0, 0, 0])

    
    plt.title("Pie Chart of Programming Languages")


    plt.show()

```
