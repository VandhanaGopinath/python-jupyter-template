### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y = [10, 5, 8, 12, 7]


if not x or not y:
    print("Error: Data is missing!")
elif len(x) != len(y):
    print("Error: Mismatch in x and y values length!")
else:
    
    plt.figure(figsize=(6, 4))
    plt.scatter(x, y, color="blue", marker="o", s=100, edgecolors="black")

    
    for i in range(len(x)):
        plt.text(x[i] + 0.1, y[i], f"({x[i]}, {y[i]})", fontsize=10, color="black")

    
    plt.title("Scatter Plot with Annotations")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True, linestyle="--", alpha=0.6)

    
    plt.show()
```
